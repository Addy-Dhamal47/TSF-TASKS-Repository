```python
new_df=df[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

## Data Exploration

```python
# Plotting Histograms to check casualties by Year, Month, Day
plt.figure(figsize=(45,20))

new_df.hist(figsize=(20,10))
```

## Terrorist Activites in Each Region through Area Plot

```python
pd.crosstab(new_df.Year, new_df.Region).plot(kind='area',figsize=(20,10))
plt.title('Terrorist Activities by Region per year')
plt.ylabel('Number of Attacks')
plt.show()
```

### Observations

- South Asia, Australasia as well as South-East Asia have a very large number of terrorist activities

### Counting Terrorist Activities each year through countplot

```python
# Plotting chart to find Terrorist attacks by year
plt.subplots(figsize=(25,15))
sns.countplot('Year',data= new_df,palette='RdYlGn_r', edgecolor=sns.color_palette("YlOrBr",10))
plt.xticks(rotation= 90)
plt.title('Number of Terrorist Activities Each Year')
plt.show()
```

### Observations

- Years 2014 and 2015 had large number of terrorist attacks .

- The frequency of the attaks gradually increased from the 1970's

### Calculating the percentage of attacks that happened since 1970 and 2017

```python
Year = new_df.Year.value_counts().to_dict()
rate=((Year[2017]-Year[1970])/Year[2017])*100
print(Year[1970],'attacks happened in 1970 ')
print(Year[2017],'attacks happened in 2017')
print('Total number of attacks from 1970 has increased by',np.round(rate,0),'% till 2017')
```

### Observations

- 2017 was the year with terrorist attacks about ~11000.

- We can also see the percentage increase in the attacks from 1970 all the way till 2017.

### Now, we will see the activites categorized by kinds of attacks

```python
# We plot a countplot to check for the method of attacks used by terrorist organizations
plt.figure(figsize=(16,8))
sns.countplot(new_df['AttackType'],order=new_df['AttackType'].value_counts().index,
              palette='BuPu_r')
plt.xticks(rotation=90)
plt.xlabel('Method')
plt.title('Method of Attack')
plt.show()
```

### Observations

Bombing as well as Explosion attacks are predominantly used by terrorist whereas Hijacking is least used .

### Now , we check the data for Different Types of Targets

```python
# We now plot to see data by different target types
plt.figure(figsize=(16,8))

sns.countplot(new_df['Target_type'],order=new_df['Target_type'].value_counts().index,
              palette='magma')
plt.xticks(rotation=90)
plt.xlabel('Type')
plt.title('Types of Targets')
plt.show()
```

### Observations

Private Citizens and Property are targeted widely by terrorist organizations as compared to other class of people

### Now, we calculate the count of terrorist attacks in each country and regions

```python
fig,axes = plt.subplots(figsize=(16,11),nrows=1,ncols=2)
sns.barplot(x = new_df['Country'].value_counts()[:20].values, y = new_df['Country'].value_counts()[:20].index, 
            ax=axes[0],palette = 'magma');
axes[0].set_title('Terrorist Attacks per Country')
sns.barplot(x=new_df['Region'].value_counts().values,y=new_df['Region'].value_counts().index,
            ax=axes[1])
axes[1].set_title('Terrorist Attacks per Region')
fig.tight_layout()
plt.show()
```

### Observations

- Frequency of terrorist attacks is more in Iraq, Pakistan and Afghanisthan.

- Attacks have more frequency in Middle East & North African as well as South Asian regions.

### Count of attacks in Each country using globe

```python
df=new_df.groupby(['Country'],as_index=False).count()
```

```python
# Countries that suffered maximum and minimum attacks


max_count=df['Year'].max()
max_id=df['Year'].idxmax()
max_name=df['Country'][max_id]
min_count=df['Year'].min()
min_id=df['Year'].idxmin()
min_name=df['Country'][min_id]
```

```python
print(max_name,'has suffered the maximum number of terror attacks of',max_count)
print(min_name,'has suffered the minimum number of terror attacks of',min_count)
```

### Finding out the most active Terrorist Organizations in the regions 

```python
# Now we check the terrorist organizations activity 
plt.subplots(figsize=(11,10))
sns.barplot(y=new_df['Group'].value_counts()[1:12].index,x=new_df['Group'].value_counts()[1:12].values,
           palette='copper')
plt.title('Most Active Terrorist Organizations')
plt.show()
```

### Analysing before and after the war( Invasion of USA in Afghanisthan in the year 2001)

```python
data_after = new_df[new_df['Year']>=2001]
fig,ax = plt.subplots(figsize=(20,20),nrows=2,ncols=1)
ax[0] = pd.crosstab(new_df.Year,new_df.Region).plot(ax=ax[0])
ax[0].set_title('Change in Regions per Year')
ax[0].legend(loc='center left',bbox_to_anchor = (1,0.5))
ax[0].vlines(x=2001,ymin=0,ymax=7000,colors='red',linestyles='--')
pd.crosstab(data_after.Year,data_after.Region).plot.bar(stacked=True,ax=ax[1])
ax[1].set_title('After Declaration of War on Terror (2001-2017)')
ax[1].legend(loc='center left',bbox_to_anchor = (1,0.5))
plt.show()
```

### Observations

- Attacks in the Middle East and North Africa have increased since 2000.

- South Asia is number 2 when it comes to terrorist attacks

### Counting Terrorist Activities by Region

```python
# we now plot to check activities by region

plt.subplots(figsize=(15,5))
sns.countplot('Region',data=new_df,palette='inferno',order=new_df['Region'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Regions')
plt.title('Number Of Terrorist Activities By Region')
plt.show()
```

### Observations

- We can see that Terrorist Activities are a lot more in Middle-Eastern, North African and South Asian regions.

- We can also observe that East and Central Asia as well as Australasia and Oceania have the least terrorist activities.

```python
d=new_df.groupby(['Year','Region'])['Wounded'].sum()
plot_df = d.unstack('Region').loc[:]
plot_df.index = pd.PeriodIndex(plot_df.index.tolist(),freq='A')
plot_df.plot(figsize=(15,8),color=sns.color_palette('Set2',12))
plt.xlabel('Year')
plt.ylabel('Wounded')
```
